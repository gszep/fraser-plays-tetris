```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lr = [0.01,0.001,0.0001]
score = [0.78,1.58,1.59]

gamma = [0.5,0.9,0.99]
gammascore = [0.84,2.0,1.58]

exploration = [1,2,3]
explorationscore = [1.5,1.9,2.0]

fig,ax = plt.subplots(1,3,figsize=(10,3))
ax[0].semilogx(lr,score, 'indianred')
ax[0].semilogx(lr,score,'o',color = 'indianred')
ax[0].set_xlabel('Learning Rate')
ax[0].set_ylabel('Average score')
ax[0].set_title('Learning Rate vs Score')

ax[1].plot(gamma,gammascore, 'indianred')
ax[1].plot(gamma,gammascore,'o',color = 'indianred')
ax[1].set_xlabel('Gamma')
ax[1].set_ylabel('Average score')
ax[1].set_title('Gamma vs Score')

ax[2].plot(exploration,explorationscore, 'indianred')
ax[2].plot(exploration,explorationscore,'o',color = 'indianred')
ax[2].set_xlabel('Amount of exporation')
ax[2].set_xticks([1,2,3])
ax[2].set_xticklabels(['short','medium','long'])
ax[2].set_ylabel('Average score')
ax[2].set_title('Exploration vs Score')




plt.tight_layout()
```
